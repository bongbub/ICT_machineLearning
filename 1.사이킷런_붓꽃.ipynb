## 사이킷런을 이용하여 붓꽃(Iris) 데이터 품종 예측하기 ##

```python
# 붓꽃 예측을 위한 사이킷런 필요 모듈 로딩
from sklearn.datasets import load_iris       # 붓꽃의 데이터 셋 가져오기

from sklearn.tree import DecisionTreeClassifier          # DecisionTreeClassifier : 결정트리분류모델을 구현한 클래스로
                                                         #                          학습시키기 위한 모델이다.

from sklearn.model_selection import train_test_split      # train_test_split : 학습데이터와 테스트데이터를 쉽게 분류할 수 있는 API
```

```python
# 데이터 세트를 로딩
iris = load_iris()          # iris 변수에 load_iris() 담음

# iris_data : iris 데이터 세트에서 피처(feature : 특징) 만으로 된 데이터를 numpy로 가지고 있다.
iris_data = iris.data       # 위 아이리스를 통해 데이터 가져오기

# iris_target : iris 데이터 셋에서 레이블(label: 결정값, 정답) 데이터를 numpy로 가지고 있다.
iris_label = iris.target    # iris_data를 통해 target 가져오기


print('iris_label => ', iris_label)                 # iris_label =>  [0 0 0 0 0 0 0 0  ....  2 2 2]  => 3개의 붓꽃의 인덱스 값 (0, 1, 2)

print('iris.타겟명 => ', iris.target_names)         # iris.타겟명 =>  ['setosa' 'versicolor' 'virginica']  => 3개의 붓꽃

# iris_label 는 인덱스로 표현한 것이고, iris.target_names은 실제 붓꽃 이름을 통해 표현한 것.

print()

# 붓꽃 데이터 세트를 자세히 보기 위해서 DataFrame으로 변환
iris_df = pd.DataFrame(data = iris_data, columns=iris.feature_names)    # 행 : iris_data , 열 : iris의 속성 명
iris_df['label'] = iris.target
iris_df.head(3)       # 컬럼 맨 끝의 label이 0  => 'setosa'
```

```python
# 1. 학습 데이터셋과 테스트 데이터셋으로 분리
```

```python
# 대문자 X : feature 데이터셋 (문제)
# 소문자 y : target 데이터셋 (정답)

X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
# 학습용 문제, 테스트용 문제, 학습용 정답, 테스트용 정답 = train_test_split(데이터, 데이터, test_size=20%, 보정치 =11)
#
```

```python
# 2. 학습데이터셋으로 학습(Train) 수행
```

```python
# DecisionTreeClassifier 객체 생성
df_dtc = DecisionTreeClassifier(random_state=11)

# 학습 수행
df_dtc.fit(X_train, y_train)      # 모델명.fit(학습용 문제, 학습용 정답)
```

```python
# 3. 테스트 데이터셋으로 예측(predict) 수행
```

```python
# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트데이터셋으로 예측 수행

pred = df_dtc.predict(X_test)        # 예측값 = 모델명.predict(테스트용문제)
```

```python
pred
# # iris.타겟명 =>  ['setosa' 'versicolor' 'virginica']  => 3개의 붓꽃 ==> [0,1,2]
```

```python
# 4. 예측 정확도 평가
```

```python
from sklearn.metrics import accuracy_score      # accuracy_score : 예측 정확도 API
print('예측정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))    # accuracy_testaccuracy_score (테스트용 정답, 예측값)

# 예측정확도 : 0.9333
```
